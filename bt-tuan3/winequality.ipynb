# BÀI LÀM: Thống kê mô tả Red Wine Quality

 ## Import thư viện 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Đọc dữ liệu


```python
df = pd.read_csv("winequality-red.csv")
```

## Xem thông tin cơ bản

```python
print("Shape:", df.shape)
print("\nCác cột dữ liệu:\n", df.columns)
print("\nThông tin tổng quát:")
print(df.info())
```

## Thống kê mô tả

```python
desc = df.describe()
print("\nThống kê mô tả:\n", desc)
```

## Kiểm tra missing values

```python
print("\nSố lượng giá trị thiếu mỗi cột:\n", df.isnull().sum())
```

## Phân tích phân phối chất lượng rượu

```python
sns.countplot(x="quality", data=df)
plt.title("Phân phối chất lượng rượu")
plt.show()
```

## Tính tương quan giữa các biến

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Ma trận tương quan giữa các thuộc tính")
plt.show()
```

## Biểu đồ boxplot cho vài biến quan trọng

```python
for col in ["alcohol", "volatile acidity", "citric acid", "density", "pH"]:
    plt.figure(figsize=(5,3))
    sns.boxplot(x="quality", y=col, data=df)
    plt.title(f"{col} theo chất lượng rượu")
    plt.show()
```
